#### Решение:

```python
df = pd.read_csv('kc_house_data.csv')
df.head()
```

*Не придумал, как сделать через условие. Но придумал формулу |x/2 - y| - x/2, после конвертировал в int64*

```python
df['delta_renovated'] = (abs(df['yr_built'] / 2 - df['yr_renovated']) - df['yr_built'] / 2)
df = df.astype({'delta_renovated': 'int64'})
```

```python
df[['yr_built', 'yr_renovated', 'delta_renovated']].head()
```

```python
df[['mth_sale', 'yr_sale']] = 0
df.head()
```

```python
df.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True)
```
