## Условие 3:

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.


Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

#### Решение:

Для начала поищем линейные зависимости:

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
```

```python
plt.figure(figsize=(10, 7))

sns.set(font_scale=1.4)

sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Correlation matrix')

plt.xticks(rotation=45);
```

Теперь можм сказать, что сильная линейная зависимость прослеживается между парами: price - sqft_living; price - grade;

И средней силы между:  price - bathrooms; price - sqft_above;

Заметим, что нет отрицательных коэфециентов. Значит нет обратных зависимостей, которые бы резко снижали стоймость жилья.

Продемонстрируем зависимости:

```python
len=4

fig, axis = plt.subplots(1, len, figsize=(21, 5), sharey=False)

def Create_figure(i: int, x: str, y: str, name: str):
    axis[i].scatter(df[y], df[x], marker='.');
    axis[i].set_xlabel(y)
    axis[i].set_ylabel(x)
    axis[i].set(title=name)

def F(l: int, x: str, y: list, name: list):

    for i in range(l):
        Create_figure(i, x, y[i], name[i])

F(l=len, x='price', y=['sqft_living', 'grade', 'bathrooms', 'sqft_above'], name=['от жилой площади', 'от кочества', 'от количества спален', 'от плщади с чердаком'])

fig.suptitle('Зависимость стоймости');
```
