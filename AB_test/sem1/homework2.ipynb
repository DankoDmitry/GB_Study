#### Задание 2.
Составьте шаблон дизайна эксперимента для гипотезы, которая набрала больше всего баллов в практическом задании предыдущего урока

##### Гипотеза

Если понизить стоймость платной подписки на месяц с 300 до 250 рублей и повысить стоймость годовой подписки с 2000 до 2400, то число платных аккаунтов увеличится на 25%, потому что некоторые люди жаловались, что платная подписка слишком дорогая.

##### Что делаем

Контрольная версия: оставляем текущие стоймости. 

Тестовая версия: предлогаем новые цены.

##### На каких пользователях тестируем: 

На случайном наборе старых пользователей без латной подписки

##### Метрики: 

Процент конверсий (%CR) — primary. Процент показателя отказов — secondary.

##### План действий:

Если наш эксперимент будет положительным, и мы зафиксируем ожидаемое улучшение в в ключевых метриках и не посадим добавочные, то масштабируем изменение и «выкатываем» его на всех пользователей.

25% подобраны так, что позволит перейти всем людям, кто пользует месечную подпску перейти на новые цены, без каких-либо убытков для компании.

Если основные метрики падают или растут недостаточно либо добавочные метрики падают, откатываем эксперимент.

В процессе проведения эксперимента может поджидать большое число нюансов. Важно понимать эти нюансы, а также способы минимизации негативного эффекта от них.
