### Увеличить доход от VK Музыки:

1. Гипотеза: 

    Конверсия платного сервиса - мала, что бы её изучить возможность её повысить, приведём гипотезу:

    Если предоставить пользователю пробный месяц бесплатного доступа, то процент пользователей возрастёт на 5%, потому что психология говорит, 
    
    что "заиметь бесплатное желает каждый, а растаться с тем, что имееешь поти никто" - йода.

2. Что делаем: Предоставляем пробный - бесплатный месяц.

    Через время ищем процент тех, кто, после истечения срока пробной версии, остался на платной подписке (K2).

3. На каких пользователях тестируем: Новые пользователи (с момента начала произведения)

4. Ключевые метрики для оценки эксперимента: Конверсия платного сервиса

    Доп. Кол. просмотров рекламы в VK Музыке.

5. Выводы: Смотрим на отношение было-стало...

    K2 > K1 + 5 - возможно имеет смысл маштабировать, иначе затраты на введение перекроют выгоду.


Контрольная версия: ничего не меняем, ищем процент тех, кто уже был и пользовался платной подпиской (K1)

Вариант протестировать гипотезу на старых неберунгах

    Действие [A] для пользователя [B] позволит увеличить метрику [C] на величину [D], потому что [E].

        А: пробный месяц бесплатного доступа - тест
        В: новые пользователи
        С: Конверсия платного сервиса
        D: 5+ (%) (предположение)
        Е: Халява (чуть более интересно описано в книге Дена Ариэли "Предсказуемая иррацианальность")
