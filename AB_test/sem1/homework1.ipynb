Какие бы ключевые метрики вы выделили? Опишите почему вы выбрали именно их и можно ли их измерить

***Выберу telegram***

##### Ключевые метрики:
1. Процент *живых* пользователей
1. Процент притока/оттока пользователей за неделю/месяц
1. Среднее число сообщений от пользователя в минуту/час/день
1. Среднее время пользование за сессию
1. Процент платных аккаунтов
1. Пиковое время пользования
1. Среднее количество занимаемой памяти на телефоне

### Задание 2. Сформулируйте 5-10 гипотез для выбранного продукта.

1. Если понизить стоймость платной подписки на месяц с 300 до 250 рублей и повысить стоймость годовой подписки с 2000 до 2400, то число платных аккаунтов увеличится на n-процентов если нет, то вернём обратно.
1. Если добавить возможность группирвать папки, то число пользователей возрастйт на n-процентов.
1. Если добавить возможность к одному аккаунту привязать несколько номеров, среднее число сообщений в минуту от одного пользователя возрастёт.
1. Если давно *молчащие* чаты будут складываться в отдельную папку, то это поможет разгрузить главную страницу, увеличит лояльность пользователя и его всемя сессии.
1. Автоматический переход на тёмную тему увеличит лояльность пользователя и его всемя сессии, возможно сдвинет пиковое время.
